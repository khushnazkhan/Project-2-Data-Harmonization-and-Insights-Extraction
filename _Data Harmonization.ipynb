```python
# outlier detection rule h y
# and after using rule calculate IQR outlier 

Q1 = final_data['atemp'].quantile(0.25)
Q3 = final_data['atemp'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1-1.5*IQR
upper_limit =Q3+1.5*IQR
```

```python
IQR
```

```python
print(lower_limit )
print(upper_limit )
```

```python
# outlier se pehle ka data h ye (shape ka use ) data ke row or column ko dekhne ke liy karte h
print('shape of dataset(before)',final_data.shape)
```

```python
# Revoving outliers

final_data = final_data[~((final_data['atemp'] < (Q1 - 1.5 * IQR)) | (final_data['atemp'] > (Q3 + 1.5 * IQR)))]

print('shape of dataset(after)',final_data.shape)
```

```python
final_data
```

```python
#capping to handle outlier

final_data['atemp']=final_data['atemp'].clip(lower=lower_limit,upper=upper_limit)
print("Data after capping")
print(final_data)
final_data['atemp']
```
